The figures in the model validation section are plotted below. 

The data for global historical population available open source from the United Nations database is utilized (data_Pop_world_un) to compare against the population from the model.

```python
fit = pd.read_csv('C:/Users/Apoorva/Downloads/fit.csv')
fit= fit.iloc[1:80, 0:12]
fit.head()
```

```python
figsize = (5,3)
ax = fit.plot(kind='scatter', x="Time", y="Pop_model", marker= 'x',color='r')    
fit.plot(kind='scatter', x="Time", y="data_Pop_world_un", marker= 's',color='b',ax=ax)    
ax.set_xlabel("Time(Years)")
ax.set_xlim(right=2025,)
ax.set_ylabel("Global Population(Hundred Thousand)")
ax.legend(["Model Output","Historical Data"])
plt.savefig('Fig3a.jpeg', dpi=300, bbox_inches='tight')

plt.show()
```

The projections for global population are available from the United Nations World Insights Report database till year 2100. They are utilized (Pop_world_med, 	Pop_world_up 	Pop_world_low) to compare against the population from the model.

```python
un = pd.read_csv('C:/Users/Apoorva/Downloads/UN comparison.csv')
un= un.iloc[0:167, 0:16]
un= un.iloc[::5, 0:16]
un.head()
```

```python
figsize = (3.5,3)
ax = un.plot(kind='line', x="Time", y="Pop", color='red')    
un.plot(kind='line', x="Time", y="Pop_world_med",  color='b',ax=ax)    
un.plot(kind='line', x="Time", y="Pop_world_up", style='--', color='dimgrey', ax=ax)    
un.plot(kind='line', x="Time", y="Pop_world_low",style='--', color='dimgrey',ax=ax)    
ax.set_xlabel("Time (Years)")
ax.set_ylabel("Global Population(Hundred Thousand)")
ax.legend(["Model projections","UN Medium Variant","UN Upper Variant","UN Lower Variant"], prop={'size': 12},fancybox=True, framealpha=0)
plt.savefig('Fig3b.jpeg', dpi = 300,  bbox_inches='tight')
plt.show()
```

The data for global carbon dioxide emissions, gross domestic product and nitrous oxide emissions is utilized (data_co2_world, 	data_GDP_world ,data_Nox_world) to compare against the respective outputs from the model.

```python
figsize = (3.5,3)
ax = fit.plot(kind='scatter', x="Time", y="co2_model", marker= 'x',color='r')    
fit.plot(kind='scatter', x="Time", y="data_co2_world", marker= 's',color='b',ax=ax)    
ax.set_xlabel("Time(Years)")
ax.set_ylabel("Carbon Emissions (kt)")
ax.set_ylim(bottom=None,
    top=4e7,)
ax.set_xlim(right=2020,)
ax.legend(["Model Output","Historical Data"], loc = 'best')
plt.savefig('Fig4a.jpeg', dpi=300, bbox_inches='tight')

plt.show()
```

```python
figsize = (5,3)
ax = fit.plot(kind='scatter', x="Time", y="GDP_model", marker= 'x',color='r')    
fit.plot(kind='scatter', x="Time", y="data_GDP_world", marker= 's',color='b',ax=ax)    
ax.set_ylim(bottom=None,
       top=10000,)
ax.set_xlim(right=2025,)
ax.set_xlabel("Time(Years)")
ax.set_ylabel("GDP(current US$)")
ax.legend(["Model Output","Historical Data"], loc='best')
plt.savefig('Fig4b.jpeg', dpi=300, bbox_inches='tight')

plt.show()
```
